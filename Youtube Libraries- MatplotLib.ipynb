### Python Matplotlib
* is a python library used for data visualisation
* can create bar, plots, scatter-plots, histograms etc

### Line Plot

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(1,11)
print(x)
```

```python
y = 2 * x
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
### Adding Titles
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show
```

```python
## changing attributes
plt.plot(x,y, color ='orange', linestyle =':', linewidth = 5)
plt.title('Line Plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show
```

```python
### Adding Two Lines in the same Plot
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
plt.plot(x,y1, color ='g', linestyle =':', linewidth = 2)
plt.plot(x,y2, color ='r', linestyle ='-', linewidth = 3)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show
```

```python
plt.subplot(1,2,1)
plt.plot(x, y1, color = 'yellow' )

plt.subplot(1,2,2)
plt.plot(x, y2, color = 'orange')

plt.show
```

### Bar Plot

```python
student = {'Bob': 87, 'Matt': 56, 'Sam': 27}
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
plt.bar(names, values)
plt.show
```

#### Adding Title and Labels

```python
plt.bar(names, values)
plt.title('Marks of Student')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(True)
plt.show
```

### Horizontal Bar Plot

```python
plt.barh(names, values, color = 'g')
plt.title('Marks of Student')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(True)
plt.show
```

### Scatter Plot
#### Creating a Basic -Scatter Plot

```python
x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
plt.scatter(x,a)
plt.show()
```

```python
### Changing Marks
plt.scatter(x,a, marker = '*', c = 'g', s = 100)
plt.show()
```

### Adding 2 markers in the same plot

```python
x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
b = [7,2,5,6,9,1,4,5,3]
plt.scatter(x,a, marker = '*', c = 'g', s = 100)
plt.scatter(x,b, marker = '.', c = 'r', s = 200)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,a,marker = '*',color = 'green', s = 100)

plt.subplot(1,2,2)
plt.scatter(x,b,marker = '.',color ='red',s= 200)
plt.show()
```

### Pie Chart

```python
fruit = ['Apple', ' Orange', ' Mango', 'Guava']
quantity = [67, 34, 100, 29]
plt.pie(quantity, labels = fruit)
plt.show
```

### Changing Attributes

```python
fruit = ['Apple', ' Orange', ' Mango', 'Guava']
quantity = [67, 34, 100, 29]
plt.pie(quantity, labels = fruit, autopct = '%0.1f%%',
       colors =['yellow', 'grey', 'blue','black'])
plt.show
```

```python
### Doughnut Chart
```

```python
fruit = ['Apple', ' Orange', ' Mango', 'Guava']
quantity = [67, 34, 100, 29]
plt.pie(quantity, labels = fruit, radius = 2, autopct = '%0.1f%%')
plt.pie([1],colors =['white'], radius = 1)
plt.show
```
