```python
import matplotlib.pyplot as plt
```

```python
# Create the histogram
plt.hist(fd['Yr'], bins= 8, alpha= 1)
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.title('Histogram of Total Assets by Date')

# Display the histogram
plt.show()
```

```python
# Create the bar graph
plt.bar(fd['Yr'], fd['Total_Assets'])
plt.xlabel('Yr')
plt.ylabel('Total_Assets')
plt.title('Bar Graph of Total Assets by Date')
plt.grid(True)
# Display the bar graph
plt.show()
```

```python
# Create the scatter plot
plt.scatter(fd['Yr'], fd['Total_Assets'])
plt.xlabel('Yr')
plt.ylabel('Total_Assets')
plt.title('Scatter Plot of Total Assets over Time')

# Display the scatter plot
plt.show()
```

```python
# Create the scatter plot
plt.scatter(fd['Yr'], fd['Revenue'])
plt.xlabel('Yr')
plt.ylabel('Revenue')
plt.title('Scatter Plot of Revenue over Time')

# Display the scatter plot
plt.show()
```

```python
# Create the scatter plots
plt.scatter(fd['Yr'], fd['Revenue'], label='Revenue')
plt.scatter(fd['Yr'], fd['Exp'], label='Expenses')
plt.xlabel('Yr')
plt.ylabel('Amount')
plt.title('Scatter Plot of Revenue and Expenses over Time')
plt.legend()

# Display the scatter plots
plt.show()
```

```python
plt.scatter(fd['Yr'], fd['Asset_Rise'])
plt.xlabel('Yr')
plt.ylabel('Asset_Rise')
plt.title('Scatter Plot of Asset_Rise over Time')

# Display the scatter plot
plt.show()
```
